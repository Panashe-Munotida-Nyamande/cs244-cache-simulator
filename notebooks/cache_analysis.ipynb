# Cache Simulator Analysis
Run simulations and plot results.

```python
import sys
sys.path.insert(0, '..')

from src.simulator import run
from src.visualiser import plot_miss_rates
import matplotlib.pyplot as plt
```

```python
# Example: sweep cache sizes on a trace
TRACE = '../traces/matrix_mult.trace'

configs = [
    {'size_bytes': 1024,  'associativity': 1, 'block_size': 64, 'label': '1KB DM'},
    {'size_bytes': 4096,  'associativity': 2, 'block_size': 64, 'label': '4KB 2-way'},
    {'size_bytes': 16384, 'associativity': 4, 'block_size': 64, 'label': '16KB 4-way'},
]

results = []
for cfg in configs:
    r = run(TRACE, cfg['size_bytes'], cfg['associativity'], cfg['block_size'])
    results.append({'label': cfg['label'], 'miss_rate': r['miss_rate']})
    print(f"{cfg['label']}: miss rate = {r['miss_rate']:.2%}")
```

```python
fig = plot_miss_rates(results)
plt.savefig('../results/miss_rates.png', dpi=150)
plt.show()
```
